```python
!pip install deepface
from deepface import DeepFace
import cv2 # Compurt Visualisation
import matplotlib.pyplot as plt # use to make a different type of graps with X,Y, co-ordinates
```

```python
img =cv2.imread("/content/img2.jpg")
result = DeepFace.analyze(img, actions = ['emotion','gender'],enforce_detection=False)
print(result)
emotion = result['dominant_emotion']
gender = result['gender'] 
print("\n"+"Emotion:" + emotion + "\n"+ "Gendar:" + gender)

face = result['region']
x = face['x']
y = face['y']
w = face['w'] + x
h = face['h'] + y

img1 =img[y:h,x:w]
plt.imshow(img1) # imshow is the kayword for presenting img in the form of plot
plt.show()
```
