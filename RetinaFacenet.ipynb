```python
from google.colab import files
!pip install retina-face
!pip install deepface
from deepface import DeepFace
from retinaface import RetinaFace
import cv2
import matplotlib.pyplot as plt
```

```python
faces = RetinaFace.extract_faces(img_path = "/content/test10.jpg", align = False)
for face in faces:
  # nam = "/content/img"+str(num)+".jpg"
  # cv2.imwrite(nam,face)
  result = DeepFace.analyze(face, actions = ['emotion','gender'],enforce_detection=False)
  emotion = result['dominant_emotion']
  gender = result['gender'] 
  print("\n"+"Emotion:" + emotion + "\n"+ "Gendar:" + gender)
  plt.imshow(face)
  plt.show()
```

```python
img = cv2.imread("/content/test12.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
if img.shape[0] < 800 or img.shape[1] < 800:
  img = cv2.resize(img, (0, 0), fx = 2, fy = 2)
resp = RetinaFace.detect_faces(img)
```

```python
l =len(resp) +1
face_loc = []
for fc in range(1,l):
  nam = "face_"+ str(fc)
  face = resp[nam]
  face_loc.append(face['facial_area'])

for (x,y,w,h) in face_loc:
  x -= 20
  y -= 20
  w += 50
  h += 50
  img1 = img[y:h,x:w]
  result = DeepFace.analyze(img1, actions = ['emotion','gender'],enforce_detection=False)
  emotion = result['dominant_emotion']
  gender = result['gender'] 
  print("\n"+"Emotion:" + emotion + "\n"+ "Gendar:" + gender)
  plt.imshow(img1) # imshow is the kayword for presenting img in the form of plot
  plt.show()
```
